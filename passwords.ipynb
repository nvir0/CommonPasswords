```python
import pandas as pd
```

```python
df = pd.read_csv('common_passwords.csv')
df
```

```python
bad_words = pd.read_csv('bad-words.csv')
bad_words
```

```python
df['is_bad']= df['password'].str.fullmatch('|'.join(bad_words['word']))
```

```python
df.loc[df['is_bad']==True]
```

```python
names = pd.read_csv('babynames-clean.csv', names=['name', 'sex'], header=None)
names['name'] = names['name'].str.lower()
names
```

```python
df['is_name'] = df['password'].str.fullmatch('|'.join(names['name']), case=False)
```

```python
df.loc[df['is_name']==True]
```

```python
df['is_date'] = df['password'].str.match('[0-3]\d[0-1]\d[1-2][0,9]\d\d')
```

```python
df.loc[df['is_date']==True]
```

```python
df.describe()
```

```python
df.to_csv('common_passwords_modified1.csv', index=False)
```
